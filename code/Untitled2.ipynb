```python
import pandas as pd

# Read the data from the file into a Pandas DataFrame
data = pd.read_csv("NSL_RQ1.txt", delimiter="\t")

# Get unique learning rates in the data
learning_rates = data["Learning Rate"].unique()

# Create a dictionary to hold the tables
tables = {}

# Loop through each learning rate and create a table for it
for lr in learning_rates:
    # Select the data for the current learning rate
    lr_data = data[data["Learning Rate"] == lr]
    
    # Drop the Learning Rate column since it is the same for all rows
    lr_data = lr_data.drop("Learning Rate", axis=1)
    
    # Set the index to be the Beta column
    lr_data = lr_data.set_index("Beta")
    
    # Rename the columns to remove spaces
    lr_data = lr_data.rename(columns={"Memory Size": "MemorySize", "AUC": "AUC", "Time": "Time"})
    
    # Store the table in the dictionary
    tables[lr] = lr_data.to_latex()

# Combine the tables side by side with the learning rate as the column header
combined_tables = "\n\n".join([f"\\textbf{{Learning Rate: {lr}}}\n\n{table}" for lr, table in tables.items()])

# Print the combined tables
print(combined_tables)

print("\\hskip 0.5in") # Add some space between tables
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# read data from file
data = pd.read_csv('NSL_RQ1.txt', delimiter='\s+',nrows=120)

# filter out lines with 'Best Parameters'
#data = data[~data['Learning'].str.contains('Best Parameters')]

# bin the AUC values into 5 categories
data['AUC_category'] = pd.cut(data['AUC'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])

# plot Memory Size Vs Beta Vs AUC for each learning rate
for lr in data['Learning'].unique():
    plt.figure()
    plt.title(f"Learning Rate: {lr}")
    for i, row in data[data['Learning']==lr].iterrows():
        plt.scatter(row['Memory'], row['Beta'], c=row['AUC_category'], cmap='viridis')
    plt.colorbar(label='AUC')
    plt.xlabel('Memory Size')
    plt.ylabel('Beta')
    plt.show()
```
